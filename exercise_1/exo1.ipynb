### part 1 : data distribution and the law of large numbers

The goal of this exercise is to manipulate a data distribution and to get familiar
with the law of large numbers in an informal way.

1) Propose a 2-dimensional random variable Z = (X, Y), with X and Y being two
real (∈ R), discrete or continuous random variables. These two variables should
represent a quantities of your choice (e.g. the age of the individuals in a population,
the color of the eyes of these individuals, ...). Compute the expected value of Z, that
must be finite.

2) Sample a number n (of your choice) of points from the law of Z and plot them
in a 2 dimensional figure.

3) For increasing values n, compute the empirical average of the first n samples
and verify that it converges to the expected value, by plotting the euclidean distance
between the empirical average and the expected value as a function of n.
Remarks :

### First let's define the variables X and Y

$$ Z = (X, Y)$$

Where:
- $X$ is the age of the individuals in a population (between 0 and 100)
- $Y$ is the monlthy income of the individuals in a population (between 0 and 10000)

Let's assume that the age and the income of the individuals in a population are uniformly distributed on floats between 0 and 100 and 0 and 10000 respectively.

So the expected value of Z is:

$$ E[Z] = (E[X], E[Y]) = (50, 5000) $$

```python
EXPECTED_Z = ((0 + 100) / 2, (0 + 10000) / 2)

print("Expected value of Z =", EXPECTED_Z)
```

### 2. Sample a number n of points from the law of Z and plot them in a 2 dimensional figure.

For this part we will consider a population of 1000 individuals.


```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000

X_LIST = np.random.uniform(0, 100, N)
Y_LIST = np.random.uniform(0, 10000, N)

plt.scatter(X_LIST, Y_LIST)
plt.xlabel("X: Age of the individual")
plt.ylabel("Y: Income of the individual")
plt.show()
```

### 3. For increasing values n, compute the empirical average of the first n samples and verify that it converges to the expected value, by plotting the euclidean distance between the empirical average and the expected value as a function of n.

```python
EMPIRICAL_Z = (np.mean(X_LIST), np.mean(Y_LIST))

print("Empirical average of Z =", EMPIRICAL_Z)

# For increasing values n, compute the empirical average of the first n samples and verify that it converges to the expected value, by plotting the euclidean distance between the empirical average and the expected value as a function of n.

euclidean_distance = lambda x, y: np.linalg.norm(np.array(x) - np.array(y))

distances = []

for n in range(1, N):
    empirical_average = (np.mean(X_LIST[:n]), np.mean(Y_LIST[:n]))
    distance = euclidean_distance(empirical_average, EXPECTED_Z)
    distances.append(distance)
plt.plot(distances)
plt.xlabel("Number of samples")
plt.ylabel("Euclidean distance of empirical average and expected Z")
plt.show()
```

# Conclusion

We can see that the empirical average converges to the expected value.
